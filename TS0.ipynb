una señal electrica de la forma sinusoidal puede ser descripta por la formula 

\begin{equation}
x(t) = vmax\cdot sen(ff\cdot 2\cdot \pi\cdot t+ph)+dc  
\end{equation}

esta formula describe el movimiento de la corriente en el tiempo, los parametros utilizados son

vmax: tencion maxima que alcanza la señal 

ff: frecuencia de la señal 

ph: la fase o desfase a la cual esta sometida la señal 

dc: el valor medio de la tencion de la señal

este ultimo parameto se da por la ecuacion de

\begin{equation}
dc = \frac{1}{T}* \int_{0}^{T} x(t) \cdot dx  
\end{equation}

siendo T el periodo de la señal. si remplazamos  x(t) por la funcion que describe el movimiento del seno nos daremos cuenta que el termino sumando ,en la funcion del movimiento, va a ser igual al valor medio de la tecion. cabe aclarar que esto solo sucede con funciones seno/coseno armonicas, en caso de tener otra hara falta su propio analizis  


### codigo 
una vez entendido como funciona la funcion y sus parametros pasaremos a escribir una funcion en spyder la cual reciba estos parametro y nos indique el movimiento en funcion de l tiempo 

```python
import numpy as np
import matplotlib.pyplot as plt

def funcion_sen(vmax,dc,ff,ph,nn,fs):
    
    t=np.arange(0,(nn+1)/fs,1/fs)
    
    y=vmax*np.sin(ff*2*np.pi*t-ph)+dc
    
    return t ,y 
```

los parametro nn y fs son, la cantidad de muestras que tenemos y la frecuencia a las que aparecen, esto hace que las muestras sean equisdistantes entre si. ademas el comando arange nos sirve para subdividir el eje x de forma que genera un vector con todos los valores del eje x los cuales son iguales a la cantidad de muestras 

```python
ff=500
ph=0
vmax=1
dc=1
nn=50
fs=25000

#para ver tantas muestras en un periodo, hay que hacer "periodo/muestras=periodo de muestro"

tiempo , espacio = funcion_sen(vmax,dc,ff,ph,nn,fs)

plt.figure(1)

plt.plot(tiempo,espacio, linestyle='-.' ,color='r')
plt.plot(tiempo,espacio, linestyle='',marker='.' ,color='b')
plt.grid(True, linestyle=':') 
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.axis([np.min(tiempo),np.max(tiempo),np.min(espacio)-0.5,np.max(espacio)+0.5]) 
plt.draw()
```

el codigo tambien nos sirve si queremos comparar distintas frecuencias de una misma señal 

```python
#caso frecuencia 500

plt.figure(2)

ff=500
ph=0
vmax=1
dc=0
nn=500
fs=100000

tiempo , espacio = funcion_sen(vmax,dc,ff,ph,nn,fs)

plt.plot(tiempo,espacio,label="500hz", linestyle='-' ,color='r')

#caso frecuencia 999

ff=999
ph=0
vmax=1
dc=0
nn=500
fs=100000

tiempo , espacio = funcion_sen(vmax,dc,ff,ph,nn,fs)

plt.plot(tiempo,espacio, label="999hz",linestyle='-' ,color='b')

#caso frecuencia 1001

ff=1001
ph=0
vmax=1
dc=0
nn=500
fs=100000

tiempo , espacio = funcion_sen(vmax,dc,ff,ph,nn,fs)

plt.plot(tiempo,espacio, label="1001hz",linestyle='-' ,color='g')

#caso frecuencia 2001

ff=2001
ph=0
vmax=1
dc=0
nn=500
fs=250000

tiempo , espacio = funcion_sen(vmax,dc,ff,ph,nn,fs)

plt.plot(tiempo,espacio, label="2001hz",linestyle='-' ,color='y')

plt.grid(True, linestyle=':') 
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.axis([np.min(tiempo),np.max(tiempo),np.min(espacio)-0.5,np.max(espacio)+0.5]) #en donde aparece el cuadro
plt.legend() 
plt.draw()


plt.tight_layout() 
plt.show()
```
