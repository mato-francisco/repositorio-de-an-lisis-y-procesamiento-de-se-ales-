## introduccion
a la hora de usar la tranformada de fourier discreta, es decir con un numero de muestras discretas, pueden llegar suceder fenomenos que son dificiles de explicar con una analisis de fourier continuo,ya que solo suceden en el mundo discreto. uno de estos fenomenos es el desparramo espectral, el cual genera que la la energia en frecuenca de una seal se desparrame en todo el espectro de frecuencia, haciendo que sea dificil poder estimar frecuencias y amplitudes de distintas seales.

este fenomeno es debido a que la tranformada de fourier en el mundo discreto hace que la seal de entrada se multiplicada por una ventana rectangular implicita de la forma
\begin{equation}
w(n)= \left\{ \begin{array}{lcc} 0 & si & n< 0 \\ 1 & si & n \geq 0 \\ 0 & si & n >N  \end{array} \right.
\end{equation}
con N siendo el numero de muestras validas.
al ventanear la seal hace que al pasarla al dominios de frecuencia,la seal de entrada en frecuencias sea convolucionada por una sinc.pero la ventana implicita no es  cualquier sinc sino que es un Karnel de dishlet en el dominio de frecuencia, la cual tiene la siguiente forma
\begin{equation}
w(k)= \frac{sin(\pi*k)}{sin(\frac{\pi*k}{N})}*\phi(k)
\end{equation}
siendo $\phi(k)$ la fase de la seal que no nos interesa mucho porque vamos a estar trabajando en modulo(y el modulo de la fase es 1), si graficamos el modulo veremos que el grafico tiene los cruzes por 0 en los enteros exepto en el 0.
 
el interpolar por zero padding hace que en la seal que le metemos a la DFT se extienda a muchas muestras, pero que valen 0, para hacer que la resolucion espectral de la seal se mucha mas grande y por ende vamos a tener un espectro de frecuencias mucho mayor por lo que seran visibles mas frecuencias que quizas no eran visibles por la resolucion espectral anterior.

## analisis y desarrollo

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft


nn=100
fs=nn
deltaf=fs/nn

t=np.arange(0,(nn)/fs,1/fs)
y=np.sin(np.pi*2*(nn/4)*t)
dev_y=np.std(y)
y=y/dev_y

z=np.sin(np.pi*2*((nn/4)+0.25)*t)
dev_z=np.std(z)
z=z/dev_z

w=np.sin(np.pi*2*((nn/4)+0.5)*t)
dev_w=np.std(w)
w=w/dev_w
```

```python
fftY=np.fft.fft(y)
fftZ=np.fft.fft(z)
fftW=np.fft.fft(w)

ffYabs=np.abs(fftY)
ffZabs=np.abs(fftZ)
ffWabs=np.abs(fftW)

plt.figure(1)
plt.clf

ff=np.arange(nn)*deltaf #escala el eje en frecuencia no normalizada por resolucion espectral
                        #para pasar de eje de muestra a frecuencia, multiplico n por la resolucion espectral
plt.plot(ff,ffYabs, label="((nn)/4)",linestyle='',marker='X' ,color='b')

plt.plot(ff,ffZabs, label="((nn/4)+0.25)",linestyle='',marker='o' ,color='r')

plt.plot(ff,ffWabs, label="((nn/4)+0.5)",linestyle='',marker='s' ,color='g')


plt.legend()
plt.grid(True, linestyle=':') 
plt.xlabel('frecuencia (HZ)')
plt.ylabel('energia[j]')
plt.xlim(0,nn/2)
#plt.axis([0,nn/2,0,10000]
plt.draw()
plt.tight_layout() 

plt.show()
```

en el grafico de arriba podemos entender de mejor manera como afecta el efecto desparramo espectral a la senoidal tranformada por fourier,a simple vista podemos ver como la energia maxima de la seal en las dos frecuencias mayores a nn/4 es reducida por un factor desconocido, pero esto conserva sentido ya que la energia total del sistema se distribuyo en lo que parece ser esas montaas verde y roja.
lo segundo que podemos pensar es el porque sucede eso.para ello tenemos que tener en cuenta el el $\Delta$f=1 es decir que tenemos graficadas frecuencias en el dominio de los naturales(no tomamos frecuencias negativas),esto es inportante para entender que, al hacer la tranformada fourier y convolucionar la senoidal por la ventana rectangular en el dominio espectral, vamos a obtener el karnel de dishlet desfasada la frecuencia de la senoidal, por lo que nuestra seal de salida es:
\begin{equation}
s(k)= \frac{sin(\pi*(k-nn/4))}{sin(\frac{\pi*(k-nn/4)}{N})}
\end{equation}
si buscamos cuando la salida es 0 obtemos que
\begin{equation}
\pi*(k-nn/4)=z*\pi  \hspace{0.5cm}con \hspace{0.5cm} z\subseteq \mathbb{Z}
\end{equation}
es decir que las soluciones son
\begin{equation}
k=z+nn/4  \hspace{0.5cm}con \hspace{0.5cm} z\subseteq \mathbb{Z}
\end{equation}
ahora podemos ver que para frecuencia enteras no va a ver desparramo ya que todos los valores de frecuencia que tenemos,es decir los $\Delta f*n$, van a ser que la seal de salida sea 0 exepto en la frecuencia.ahora que sucede con frecuencia un poco desfasadas, pues que vamos a ver un efecto de desparramo por que los $\Delta f*n$ no son las soluciones para que la seal de salida sea 0.

```python
plt.plot(ff,20*np.log10(ffYabs), label="dB((nn)/4)",linestyle='',marker='X' ,color='b')

plt.plot(ff,20*np.log10(ffZabs), label="dB((nn/4)+0.25)",linestyle='',marker='o' ,color='r')

plt.plot(ff,20*np.log10(ffWabs), label="dB((nn/4)+0.5)",linestyle='',marker='s' ,color='g')


plt.legend()
plt.grid(True, linestyle=':') 
plt.xlabel('frecuencia (HZ)')
plt.ylabel('energia[db]')
plt.xlim(0,nn/2)
#plt.axis([0,nn/2,0,10000]
plt.draw()
plt.tight_layout() 

plt.show()
```

el anterior grafico muestra como el efecto de desparramo afecta a la densidad espectral de potencia. el fenomeno como habiamos dicho anteriormente desparrama la energia a todas las frecuencias, lo que genera algo a lo que podriamos llamar un tipo de ruido que saca potencia a nuestra seal. esto se ve en el grafico como el piso de ruido de la senoidal de frecuencia nn/4(seal azul) distanciado  250 escalas de magnitud con respecto a la del piso de ruido de la seales de nn/4 +0.25(roja) y nn/4 +0.5(verde). esto hace que a la hora de buscar la amplitud de una seal desconocida que tenga AWG(ruido blanco aditivo) se complique mas la busqueda.

```python
energia_y=np.sum(np.abs(y)**2)
energia_ffy=np.sum(np.abs(fftY)**2)/nn
print("energia de la seal de frecuencia N/4")
print(f"{energia_y:.2f}")
print("energia de tranformada de la seal de frecuencia N/4")
print(f"{energia_ffy:.2f}")
energia_z=np.sum(np.abs(z)**2)
energia_ffz=np.sum(np.abs(fftZ)**2)/nn
print("energia de la seal de frecuencia N/4")
print(f"{energia_z:.6f}")
print("energia de tranformada de la seal de frecuencia N/4+0.25")
print(f"{energia_ffz:.6f}")
energia_w=np.sum(np.abs(w)**2)
energia_ffw=np.sum(np.abs(fftW)**2)/nn
print("energia de la seal de frecuencia N/4+0.5")
print(f"{energia_w:.2f}")
print("energia de tranformada de la seal de frecuencia N/4+0.5")
print(f"{energia_ffw:.2f}")
```

los anterirores resultados fueron obtenidos por la identidad parsebal. nos indica que a pesar que la DFT o FFT muliplique implicitamente nuestra seal de entrada por una ventana rectangular, la energia de la seal de entrada no se vera afectada,esto podria llegar a ser poco intuitivo ya que multiplicar por una ventana hace que la seal de entrada pierda energia.
ademas no importa a la frecuencia a la que este la seal de entrada,parsebal se seguira cumpliendo.
